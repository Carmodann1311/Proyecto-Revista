```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos el rango de precios
p = np.linspace(0, 20, 100)

# Funciones de demanda y oferta
D = 1000 - 50 * p  # Demanda
S = 200 + 30 * p    # Oferta

# Creamos la gráfica
plt.figure(figsize=(10, 6))
plt.plot(p, D, label='Demanda (D)', color='blue')
plt.plot(p, S, label='Oferta (S)', color='orange')

# Punto de equilibrio
p_eq = 10
q_eq = 500
plt.plot(p_eq, q_eq, 'ro')  # Punto de equilibrio
plt.annotate(f'Equilibrio\n(${p_eq}, {q_eq})', xy=(p_eq, q_eq), xytext=(p_eq+2, q_eq-100),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Etiquetas y título
plt.title('Oferta y Demanda de PlayStation 5')
plt.xlabel('Precio ($)')
plt.ylabel('Cantidad')
plt.xlim(0, 20)
plt.ylim(0, 1200)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```
